### Libraries

```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime as dt
```

### Dataset

```python
stream = pd.read_csv('stream.csv')
stream.head()
```

### Bar Graph: Featured Games

```python
viewers_df = pd.DataFrame(stream.groupby('game').login.count().reset_index())
viewers_df.rename(columns={'login': 'count'}, inplace=True)
viewers_df = viewers_df[viewers_df['count'] > 10000].sort_values(by='count', ascending=False)
viewers_df
```

```python
games = list(viewers_df['game'])
games
```

```python
viewers = list(viewers_df['count'])
viewers 
```

```python
plt.figure(figsize=(8,6))
ax = plt.subplot(1,1,1)
plt.bar(x=range(len(games)), height=viewers, tick_label= range(len(games)), color='slateblue')
ax.set_xticklabels(games)
plt.xticks(rotation=30, ha='right')
plt.legend(["Twitch"])
ax.set_facecolor("lightgrey")
plt.title('Number of Viewers per Game', size = 18)
plt.show()
```

### Pie Chart: League of Legends Viewers' Whereabouts

```python
df_label = stream[stream['game'] == 'League of Legends'].groupby('country').login.count().dropna().reset_index()
df_label.rename(columns={'login': 'count'}, inplace=True)
df_label = df_label[df_label['count'] > 5000]
df_label
```

```python
labels = list(df_label['country'])
labels
```

```python
countries = list(df_label['login'])
countries
```

```python
colors = ['lightskyblue', 'gold', 'lightcoral', 'darkseagreen']
explode = (0, 0, 0, 0.1)
plt.figure(figsize=(8,6))
plt.pie(x=countries, explode=explode, colors=colors, shadow=True, startangle=320, autopct='%1.0f%%', pctdistance=1.15)
plt.axis('equal')
plt.legend(labels, loc='upper left')
plt.title('League of Legends Viewers by Country', size=16)
plt.show()
```

### Line Graph: Time Series Analysis

```python
stream['time'] = pd.to_datetime(stream['time'])
time_df = pd.DataFrame(stream.groupby(stream['time'].dt.hour).login.count().reset_index())
time_df.rename(columns={'login': 'count'}, inplace=True)
time_df
```

```python
hour = range(24)
viewers_hour = list(time_df['count'])
```

```python
y_upper = [i + (i*0.15) for i in viewers_hour]
y_lower = [i - (i*0.15) for i in viewers_hour]
plt.figure(figsize=(10,6))
ax = plt.subplot(1,1,1)
plt.plot(hour, viewers_hour)
plt.xticks(hour)
plt.xlabel('Hour', size = 14)
plt.ylabel('Count', size = 14)
plt.legend(['2015-01-01'])
plt.title('Number of Viewers per Hour', size = 18)
plt.fill_between(hour, y_lower, y_upper, alpha=0.2)
ax.set_facecolor("gainsboro")
plt.show()
```
